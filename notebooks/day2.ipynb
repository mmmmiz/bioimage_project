# 最小動線（読み込み→カラー表示）だけ実行

```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
```

```python
BASE_DIR = Path.cwd().parent
path = BASE_DIR / "data" / "sample.jpg"
```

今いるパス（notebook）の上（プロジェクトディレクトリ）をBASE_DIRにして、
そこから画像のパスを指定

```python
img_bgr = cv2.imread(str(path), cv2.IMREAD_COLOR)
if img_bgr is None:
    raise ValueError(f"画像の読み込みに失敗: {path}")
```

OpenCVで読み込んだ画像（img_bgr）はBGRで読み込まれている

```python
print("Color(BGR) shape:", img_bgr.shape)
print("Color(BGR) dtype:", img_bgr.dtype)

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.figure()
plt.title("Color (RGB)")
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

cvtColorでBGRからRGBに変更してmatplotlibで描画
※ matplotlibは RGB
