```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
BASE_DIR = Path.cwd().parent
path = BASE_DIR / "data" / "sample.jpg"
```

```python
img_gray = cv2.imread(str(path), cv2.IMREAD_GRAYSCALE)
if img_gray is None:
    raise ValueError(f"画像の読み込みに失敗: {path}")
```

```python
plt.figure()
plt.title("img_gray")
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.show()
```

```python
def calc_contrast(img_gray) -> float:
    """
    グレースケール画像のコントラストを計算する
    コントラスト = 画素値の標準偏差
    """
    return float(np.std(img_gray))
contrast = calc_contrast(img_gray)
print("contrast", contrast)
```
